# Iris

```python
iris = load_iris_to_df()
iris_label = iris.get('label')
print(iris_label.shape)
iris = iris.drop('label', 1)
print(iris.shape)
pca_iris, _ = reduce_dimension(iris, 'PCA', 4,plot_curve=True,title='PCA - Iris - Variance')
pca_iris, _ = reduce_dimension(iris, 'PCA', 2)

ica_iris, _ = reduce_dimension(iris, 'ICA', 4,plot_curve=True,title='ICA - Iris - kurtosis (sorted)')
ica_iris, _ = reduce_dimension(iris, 'ICA', 2,plot_curve=True,ica_sort=False)

ica_iris,_=reduce_dimension(iris, 'ICA',2)
rnd_iris, _ = reduce_dimension(iris, 'random', 2)
fa_iris, _ = reduce_dimension(iris, 'FA', 2, labels=iris_label)
```

```python
plot_2d(pca_iris.assign(label=iris_label), "0", "1", comment="pca_iris",title=" IRIS - distribution of original label after PCA")
plot_2d(ica_iris.assign(label=iris_label), "0", "1", comment="ica_iris",title=" IRIS - distribution of original label after ICA")
plot_2d(rnd_iris.assign(label=iris_label), "0", "1", comment="rnd_iris",title=" IRIS - distribution of original label after Random")
plot_2d(fa_iris.assign(label=iris_label), "0", "1", comment="fa_iris",title=" IRIS - distribution of original label after FA")
```

```python
kmeans_pca_iris = clustering(pca_iris, 'kmeans', 3)
em_pca_iris = clustering(pca_iris, 'EM', 3)
plot_2d(kmeans_pca_iris, "0", "1", comment="pca_iris_kmeans",title='After PCA - clustering of Iris by K-means')
plot_2d(em_pca_iris, "0", "1", comment="pca_iris_EM",title='After PCA - clustering of Iris by EM')
```

```python
kmeans_ica_iris = clustering(ica_iris, 'kmeans', 3)
em_ica_iris = clustering(ica_iris, 'EM', 3)
plot_2d(kmeans_ica_iris, "0", "1", comment="ica_iris_kmeans",title='After ICA - clustering of Iris by K-means')
plot_2d(em_ica_iris, "0", "1", comment="ica_iris_EM",title='After ICA - clustering of Iris by EM')
```

```python
kmeans_rnd_iris = clustering(rnd_iris, 'kmeans', 3)
em_rnd_iris = clustering(rnd_iris, 'EM', 3)
plot_2d(kmeans_rnd_iris, "0", "1", comment="rnd_iris_kmeans",title='After RND - clustering of Iris by K-means')
plot_2d(em_rnd_iris, "0", "1", comment="rnd_iris_EM",title='After RND - clustering of Iris by EM')
```

```python
kmeans_fa_iris = clustering(fa_iris, 'kmeans', 3)
em_fa_iris = clustering(fa_iris, 'EM', 3)
plot_2d(kmeans_fa_iris, "0", "1", comment="f_iris_kmeans",title='After FA - clustering of Iris by K-means')
plot_2d(em_fa_iris, "0", "1", comment="f_iris_EM",title='After FA - clustering of Iris by EM')
```
